```python
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

st.title("Instructor Performance and Course Quality Evaluation")

# Load Data
teachers = pd.read_excel("Instructor_Data.xlsx.xlsx", sheet_name="Teachers")
courses = pd.read_excel("Instructor_Data.xlsx.xlsx", sheet_name="Courses")
transactions = pd.read_excel("Instructor_Data.xlsx.xlsx", sheet_name="Transactions")

# Merge Data
df = pd.merge(transactions, teachers, on="TeacherID")
df = pd.merge(df, courses, on="CourseID")

# KPIs
st.header("Key Performance Indicators")

col1, col2 = st.columns(2)

with col1:
    st.metric("Average Teacher Rating",
              round(df["TeacherRating"].mean(), 2))

with col2:
    st.metric("Average Course Rating",
              round(df["CourseRating"].mean(), 2))

# Expertise Filter
expertise = st.selectbox(
    "Select Expertise",
    df["Expertise"].unique()
)

filtered = df[df["Expertise"] == expertise]

st.write(filtered.head())

# Scatter Plot
st.subheader("Experience vs Teacher Rating")

fig, ax = plt.subplots()

sns.scatterplot(
    x="YearsOfExperience",
    y="TeacherRating",
    data=filtered,
    ax=ax
)

st.pyplot(fig)

# Top Teachers
st.subheader("Top Teachers")

top_teachers = df.groupby(
    "TeacherName"
)["TeacherRating"].mean().sort_values(
    ascending=False
).head(10)

st.bar_chart(top_teachers)
```
